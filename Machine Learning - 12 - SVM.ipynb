```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import svm
```

### With Hill-Valley Data

```python
## Importing the data
X_train = np.loadtxt('C:/Users/Sourita Nag/OneDrive/Documents/Fourth Semester/hill-valley/X.dat')
X_train.shape

Y_train = np.loadtxt('C:/Users/Sourita Nag/OneDrive/Documents/Fourth Semester/hill-valley/Y.dat')
Y_train.shape

X_test = np.loadtxt('C:/Users/Sourita Nag/OneDrive/Documents/Fourth Semester/hill-valley/Xtest.dat')
X_test.shape

Y_test = np.loadtxt('C:/Users/Sourita Nag/OneDrive/Documents/Fourth Semester/hill-valley/Ytest.dat')
Y_test.shape
```

### Part a

```python
##Fitting the Classifier in sklearn
miscl_train_all = []
miscl_test_all = []
d = np.arange(1,16,1)

for val in d:
    
    clf = svm.SVC(C = 1,kernel='poly',degree = val)
    clf.fit(X_train, Y_train)
    Y_pred_train = clf.predict(X_train)
    misscl_error_train = np.mean(Y_train!=Y_pred_train)
    miscl_train_all.append(misscl_error_train)
    Y_pred_test = clf.predict(X_test)
    misscl_error_test = np.mean(Y_test!=Y_pred_test)
    miscl_test_all.append(misscl_error_test)
plt.plot(d,miscl_train_all,label = 'Train')
plt.plot(d,miscl_test_all,label = 'Test')
plt.xlabel('Polynomial Degrees')
plt.ylabel('Missclassification Errors')
plt.legend()
plt.show()
```

### Part b

```python
d = np.arange(1,22,1)
miscl_train_all = []
miscl_test_all = []

for val in d:
    g = 1/(2**((val-1)))
    clf = svm.SVC(C = 1,kernel='rbf',gamma = g)
    clf.fit(X_train, Y_train)
    Y_pred_train = clf.predict(X_train)
    misscl_error_train = np.mean(Y_train!=Y_pred_train)
    miscl_train_all.append(misscl_error_train)
    Y_pred_test = clf.predict(X_test)
    misscl_error_test = np.mean(Y_test!=Y_pred_test)
    miscl_test_all.append(misscl_error_test)
plt.semilogx(d, miscl_train_all, marker = ".",label = 'Train')
plt.semilogx(d, miscl_test_all, marker = "*",label = 'Test')
plt.legend()
plt.xlabel('Polynomial Degrees')
plt.ylabel('Missclassification Errors')
plt.show()
```
