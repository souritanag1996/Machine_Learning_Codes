```python
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,roc_curve,roc_auc_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[11,9]
eps=1e-05
gisette_train=pd.read_csv("Gisette/gisette_train.data",delimiter=" ",header=None)
print(gisette_train.shape)
gisette_train=gisette_train.drop(5000,axis=1) # 
print('new shape is',gisette_train.shape)
gisette_valid=pd.read_csv("Gisette/gisette_valid.data",delimiter=" ",header=None)
print(gisette_valid.shape)
gisette_valid=gisette_valid.drop(5000,axis=1) #
print('new shape is',gisette_valid.shape)
gisette_label_train=pd.read_csv("Gisette/gisette_train.labels",delimiter=" ",header=None)
gisette_label_valid=pd.read_csv("Gisette/gisette_valid.labels",delimiter=" ",header=None)
# Normalization
scalar= StandardScaler()
scalar.fit(gisette_train)
X_train_normalized=scalar.transform(gisette_train)
X_valid_normalized=scalar.transform(gisette_valid)
# Adding bias 
# X_train=pd.DataFrame(np.hstack([np.ones((X_train_normalized.shape[0],1)),X_train_normalized]))
# print('final shape is',X_train.shape)
# X_test=pd.DataFrame(np.hstack([np.ones((X_valid_normalized.shape[0],1)),X_valid_normalized]))
# print('final shape is',X_test.shape)
X_train=pd.DataFrame(X_train_normalized)
X_test=pd.DataFrame(X_valid_normalized)
y_train=gisette_label_train
y_test=gisette_label_valid
y_train.loc[y_train[0]==-1,:]=0
y_test.loc[y_test[0]==-1,:]=0
df=pd.DataFrame(np.hstack([np.ones((X_train.shape[0], 1)), X_train]))
df_test=pd.DataFrame(np.hstack([np.ones((X_test.shape[0], 1)), X_test]))
def loss(y,h):
    return(np.log(1 + np.exp(-(2*y-1)*h)))    
```

```python
table={'K':[],'Train Misclassification':[],'Test Misclassification':[],'feature selected':[]}
#def logitboost(X_train,X_test,boost_itr,weights):
df=pd.DataFrame(np.hstack([np.ones((X_train.shape[0], 1)), X_train]))
df_test=pd.DataFrame(np.hstack([np.ones((X_test.shape[0], 1)), X_test]))
h=np.zeros(X_train.shape[0])
w=np.zeros(X_train.shape[1]+1)
k=501#30,100,300,500
val=[10,30,100,300,500]
final_log_loss=[]
for k_val in range(1,k):
    px=1/(1+np.exp(-h))
    #px=np.maximum(px, 2. * eps)
    #print(px)
    #z=(y_train[0]-px)/(px*(1-px))
    z=np.where(y_train[0],1./px, -1./(1. - px))
    z = np.clip(z, a_min=-4, a_max=4)
    #print(z)
    W=px*(1-px)
    # calculating b1 vector
    beta1=[]
    beta0=[]
    h_x=[]
    for i in range(0,X_train.shape[1]):
        x_bar=np.sum(W*X_train[i])/np.sum(W)
        z_bar=np.sum(W*z)/np.sum(W)
        b1=np.sum(W*(z-z_bar)*(X_train[i]-x_bar))/np.sum(W*(X_train[i]-x_bar)**2)
        b0=z_bar-b1*x_bar
        h_i=h+b0+b1*X_train[i]
        beta1.append(b1)
        beta0.append(b0)
        h_x.append(h_i)
    log_loss=[]
    for i in range(0,len(h_x)):
        log_loss.append(np.mean(loss(y=y_train[0],h=h_x[i])))
    m=[i for i, e in enumerate(log_loss) if e == min(log_loss)]
    print(m)
    j=[i for i, e in enumerate(log_loss) if e == min(log_loss)][0]
    beta1_final=beta1[j]
    w[j+1]+=beta1_final
    print(beta1_final)
    w[0]+=beta0[j]
    h=np.dot(df,w)
    if k_val in val:
        # Making predictions on training data
        y_train_pred=1/(1+np.exp(-h))
        y_train_pred_label=np.where(y_train_pred>0.5,1,0)
        train_error=1-accuracy_score(y_train,y_train_pred_label)
        # Making predictions on testing data    
        h_t=np.dot(df_test,w)
        y_test_pred=1/(1+np.exp(-h_t))
        y_test_pred_label=np.where(y_test_pred>0.5,1,0)
        test_error=1-accuracy_score(y_test,y_test_pred_label)
        table['K'].append(k_val)
        table['Train Misclassification'].append(train_error)
        table['Test Misclassification'].append(test_error)
        table['feature selected'].append(np.count_nonzero(w))
        print(pd.DataFrame(table))
    final_log_loss.append(np.mean(loss(y=y_train[0],h=h)))
    print("currently running",k_val," iteration")

    if k_val==100:
        fpr_train,tpr_train,_=roc_curve(y_train,y_train_pred)
        fpr_test,tpr_test,_=roc_curve(y_test,y_test_pred)
        train_auc=roc_auc_score(y_train,y_train_pred)
        test_auc=roc_auc_score(y_test,y_test_pred)
        plt.plot(fpr_train,tpr_train,label='Training (AUC={:.3f})'.format(train_auc))
        plt.plot(fpr_test,tpr_test,label='Testing (AUC={:.3f})'.format(test_auc))
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.legend()
        plt.title("ROC curve for gisette dataset")
        plt.show()
    
    if k_val==500:
        plt.plot(range(1,501),final_log_loss)
        plt.title("Loss Vs iteration graph")
        plt.xlabel('iteration')
        plt.ylabel('Loss')
        plt.show()

```

```python
table=pd.DataFrame(table)
table.to_excel('hw6_gisette.xlsx')
table
```
